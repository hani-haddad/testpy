```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
Url='https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation='
```

```python
page=requests.get(Url)
page
```

```python
soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
scraped_jobOffers=soup.find_all('li',class_="clearfix job-bx wht-shd-bx")
comp_name =scraped_jobOffers[0].find('h3',class_="joblist-comp-name").string.replace("\r\n","").strip(" ")
comp_name
scraped_jobOffers[0]
```

```python
comp_names=[]
skills=[]
publishing_date=[]
experiences=[]
location=[]
description=[]
job_links=[]
for job in scraped_jobOffers :
  timeOfpublishing = job.find('span',class_="sim-posted").text.replace("\n","")
  comp_name =job.find('h3',class_="joblist-comp-name").text.replace("\r\n","").strip(" ")
  KeySkills = job.find('span',class_="srp-skills").text.replace("\r","").replace("\n","").strip(" ")
  experience = job.li.contents[1]
  job_location = job.find_all("i","material-icons")[-1].next_sibling.next_sibling.text
  job_description = job.find('ul',class_="list-job-dtl clearfix").li.text.replace("\r\n","").replace("\n","")[16:]
  job_link= job.a.attrs['href']

  comp_names.append(comp_name)
  skills.append(KeySkills)
  publishing_date.append(timeOfpublishing)
  experiences.append(experience)
  location.append(job_location)
  description.append(job_description)
  job_links.append(job_link)
```

```python
data = pd.DataFrame()
data["comp_names"]= comp_names
data["job_location"]= location
data["Key_Skills"]= skills
data["publishing_date"]= publishing_date
data["experience"]= experiences
data["job_description"]= description
data["job_link"]= job_link

data.head(25) 
```
