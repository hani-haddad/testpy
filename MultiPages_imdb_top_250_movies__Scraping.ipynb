##For multiple pages

```python
pageNums = np.arange(1,1000,100)
pageNums
```

```python
movies_names = []
releaseYears = []
durations = []
ratings = []
metascors =[]
votes=[]
grosses =[]
```

```python
for pageNum in pageNums:

  Url="https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(pageNum)+"&ref_=adv_nxt"
  page=requests.get(Url)
  soup = BeautifulSoup(page.content,'html.parser')
  scraped_movies=soup.find_all('div',class_="lister-item-content")
  for movie in scraped_movies:
    movie_name=movie.find('h3',class_="lister-item-header").contents[3].text
    movies_names.append(movie_name)

    releaseYear= movie.find('h3',class_="lister-item-header").contents[5].text[1:5]
    releaseYears.append(releaseYear)

    duration = movie.find('span',class_="runtime").text
    durations.append(duration)

    rating = movie.find('span',class_="global-sprite rating-star imdb-rating").next_sibling.next_sibling.text
    ratings.append(rating)

    meta = movie.find('div',class_="inline-block ratings-metascore").text[1:3] if movie.find('div',class_="inline-block ratings-metascore") else "--"
    metascors.append(meta)

    vote= movie.find_all('span',{'name':'nv'})[0].text if movie.find_all('span',{'name':'nv'})[0].text else "--"
    votes.append(vote)

    e=movie.find_all('span',{'name':'nv'})

    if len(e) >= 2 :
        gross = e[1].text
        if "M" in gross:
          gross= gross
        else : 
          gross= "--"
    else : 
          gross= "--"
    
    grosses.append(gross)
```

```python
data = pd.DataFrame()
data["movie"]= movies_names
data["rating"]=ratings
data["releaseYear"]= releaseYears
data["Duration"]=durations
data["metascor"]= metascors
data["votes"]= votes
data["Gross"]= grosses
data.head(25) 
```
